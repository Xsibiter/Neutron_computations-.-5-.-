```python
import numpy as np

def elliptic_integral_monte_carlo(k, n = 1000):

    theta = np.random.uniform(0, np.pi / 2, n)
    

    integrand_values = 1 / np.sqrt(1 - k**2 * np.sin(theta)**2)
    

    integral_value = np.mean(integrand_values) * (np.pi / 2)
    return integral_value

k = 0.5
result = elliptic_integral_monte_carlo(k)
print("Значение эллиптического интеграла 1-го рода для k =", k, "равно", result)
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 100
m = 100

integral = np.zeros(m)

a = 0
b = np.pi / 2

def f(theta):
    return 1 / np.sqrt(1 - 0.5**2 * np.sin(theta)**2)

for i in range(m):
    j = 0
    theta = np.random.uniform(0,1,n) * b
    for t in range(n):
        j += f(theta[t])
    integral[i] = (b - a) * j / n
    std_values = np.zeros(m)
for i in range(1, m + 1):
    std_values[i - 1] = np.std(integral[:i]) 

plt.plot(range(40, m + 1), std_values[39:], marker='o')
plt.xlabel('параметр i')
plt.ylabel('Стандартное отклонение')
plt.title('Зависимость стандартного отклонения интеграла от значения i')
plt.show()

print("Значение эллиптического интеграла первого рода для k = 0.5 равно:", f"{np.average(integral)} +- {np.std(integral)}")

```
